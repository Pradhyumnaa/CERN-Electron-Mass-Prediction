```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the original data from "dielectron.csv"
original_data = pd.read_csv('dielectron.csv')
```

```python
# Drop rows with NaN values
original_data = original_data.dropna()
```

```python
# Calculate the number of samples for each set
total_samples = len(original_data)
validation_samples = int(0.15 * total_samples)
test_samples = int(0.05 * total_samples)
train_samples = total_samples - validation_samples - test_samples

# Randomly select the validation samples
validation_data = original_data.sample(n=validation_samples)
original_data = original_data.drop(validation_data.index)

# Randomly select the test samples
test_data = original_data.sample(n=test_samples)
original_data = original_data.drop(test_data.index)

# Save the data to separate CSV files
validation_data.to_csv('validation.csv', index=False)
test_data.to_csv('test.csv', index=False)
original_data.to_csv('training.csv', index=False)
```

```python
# Load the original data from "dielectron.csv"
training_data = pd.read_csv('training.csv')
```

```python
# Extract the 'M' column
m_values = training_data['M']

# Plot a histogram
plt.hist(m_values, bins=30, edgecolor='k')
plt.xlabel('M')
plt.ylabel('Frequency')
plt.title('Histogram of M')
plt.show()
```

```python
# Calculate the MAD
median = np.median(training_data['M'])
mad = np.median(np.abs(training_data['M'] - median))

print("Median Absolute Deviation (MAD):", mad)
```
